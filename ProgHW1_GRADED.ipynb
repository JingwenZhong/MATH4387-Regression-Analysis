        $$ H_0: \beta _0 = 2 $$
        $$ H_a: \beta _0 \ne 2  $$

c. What is the P-value of your test in part (b)? How does it support the conclusion reached in part (b)?

    $$P-value = 0.002916603883063615$$

```python
print('5#a')
print()
data=np.loadtxt('GPA.txt')
x=data[:,1]
y=data[:,0]
x=sm.add_constant(x)
results=sm.OLS(y,x).fit()
x=data[:,1]
n=y.size
y_hat=results.params[0]+results.params[1]*x
MSE=(np.sum((y-y_hat)**2))/(y.size-2)
s2_b1=MSE/np.sum((x-x.mean())**2)
s_b1=np.sqrt(s2_b1)
b1=results.params[1]
print('estimated sigma for b1 is:', s_b1)
print('The 99 percent confidence interval is : [', b1-sp.stats.t.ppf(0.995,n-2)*s_b1,b1+sp.stats.t.ppf(0.995,n-2)*s_b1,']')
print( )

print('5#b')
print()
tt=b1/s_b1
a=0.01
if abs(tt)<=sp.stats.t.ppf(0.995,n-2):
     print('conclude H0:no linear association exists'); 
elif(abs(tt)>sp.stats.t.ppf(0.995,n-2)):
    print('conclude Ha:has a linear association exists'); 
print()
    
print('5#c')
print()
print('P-value =', sp.stats.t.sf(tt, n-2)*2)
```
